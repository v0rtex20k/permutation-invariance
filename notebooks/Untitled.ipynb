```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 10})
# PyTorch
import torch
```

```python
experiments_dir = "/cluster/tufts/hugheslab/eharve06/permutation-invariance/experiments/imagenet1k"

ncols, nrows = 3, 5
fig, axs = plt.subplots(figsize=(4*ncols, 3*nrows), ncols=ncols, nrows=nrows, sharey="row")

for idx, num_perms in enumerate([1, 10, 100, 1000, 10000]):

    #batch_size, num_alphas, num_batches, num_permutations, random_state = 128, 11, 10, 100, 1001
    batch_size, num_alphas, num_batches, random_state = 128, 11, 10, 1001
    filename = f"batch_size={batch_size}_num_alphas={num_alphas}_num_batches={num_batches}_num_perms={num_perms}_random_state={random_state}.pth"
    results = torch.load(f"{experiments_dir}/{filename}", map_location=torch.device("cpu"), weights_only=False)

    for batch_idx in range(num_batches):
        axs[idx,0].plot(range(num_alphas), results["val_losses"][batch_idx], color="#1F77B4")
    axs[idx,0].set_title(f"{num_perms} permutation(s)")
    axs[idx,0].set_xticks([0, num_alphas-1])
    axs[idx,0].set_xticklabels([r"$\theta$", r"$\theta_\pi$"])
    axs[idx,0].set_ylabel("Validation NLL")

    #batch_size, num_alphas, num_batches, num_permutations, random_state = 128, 11, 10, 100, 2001
    batch_size, num_alphas, num_batches, random_state = 128, 11, 10, 2001
    filename = f"batch_size={batch_size}_num_alphas={num_alphas}_num_batches={num_batches}_num_perms={num_perms}_random_state={random_state}.pth"
    results = torch.load(f"{experiments_dir}/{filename}", map_location=torch.device("cpu"), weights_only=False)

    for batch_idx in range(num_batches):
        axs[idx,1].plot(range(num_alphas), results["val_losses"][batch_idx], color="#1F77B4")
    axs[idx,1].set_title(f"{num_perms} permutation(s)")
    axs[idx,1].set_xticks([0, num_alphas-1])
    axs[idx,1].set_xticklabels([r"$\theta$", r"$\theta_\pi$"])

    #batch_size, num_alphas, num_batches, num_permutations, random_state = 128, 11, 10, 100, 3001
    batch_size, num_alphas, num_batches, random_state = 128, 11, 10, 3001
    filename = f"batch_size={batch_size}_num_alphas={num_alphas}_num_batches={num_batches}_num_perms={num_perms}_random_state={random_state}.pth"
    results = torch.load(f"{experiments_dir}/{filename}", map_location=torch.device("cpu"), weights_only=False)

    for batch_idx in range(num_batches):
        axs[idx,2].plot(range(num_alphas), results["val_losses"][batch_idx], color="#1F77B4")
    axs[idx,2].set_title(f"{num_perms} permutation(s)")
    axs[idx,2].set_xticks([0, num_alphas-1])
    axs[idx,2].set_xticklabels([r"$\theta$", r"$\theta_\pi$"])

fig.tight_layout()
plt.show()
```
